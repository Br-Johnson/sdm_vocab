```
%load_ext rpy2.ipython
import pandas as pd
```

```
# Pass Python objects to R
import numpy as np
from rpy2.robjects.packages import importr
base = importr('base')
py_list = [1, 2, 3, 4, 5]
r_list = base.as_list(py_list)
print(r_list)
```

```
# Pass R objects to Python
from rpy2.robjects import r
r_vector = r('c(1,2,3,4,5)')
py_vector = np.array(r_vector)
print(py_vector)
```

```
# Pass a Python DataFrame to R
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
from rpy2.robjects import pandas2ri
pandas2ri.activate()
r_df = pandas2ri.py2ri(df)
print(r_df)
```
